```python
import pandas as pd
from rapidfuzz import process, fuzz
import re

# Fungsi untuk membersihkan harga (hanya angka setelah 'Rp')
def bersihkan_harga(harga):
    if pd.isnull(harga):
        return None

    # Konversi ke string dan mencari angka setelah 'Rp'
    harga_str = str(harga)

    # Cari angka yang muncul setelah 'Rp' dan pastikan kita menangkap titik sebagai pemisah ribuan
    match = re.search(r'Rp\s*(\d{1,3}(?:[.,]\d{3})*)', harga_str)

    if match:
        # Ambil angka yang ditemukan dan hapus titik (untuk pemisah ribuan) dan koma jika ada
        angka = match.group(1).replace(".", "").replace(",", "")
        return float(angka)

    return None

# Ambil semua harga bekas (tidak mengagregasi)
df_bekas = df_bekas.copy()
df_bekas['Harga'] = df_bekas['Harga'].apply(lambda x: bersihkan_harga(x))  # Bersihkan harga bekas

# Rename kolom harga bekas
df_bekas = df_bekas.rename(columns={'Harga': 'Harga_Bekas'})

# Ambil harga baru tertinggi per mobil dan tahun
df_baru = df_baru.copy()
df_baru['Harga'] = df_baru['Harga'].apply(lambda x: bersihkan_harga(x))  # Bersihkan harga baru

df_baru_agg = df_baru.groupby(['Mobil', 'Tahun'], as_index=False).agg({'Harga': 'max'})
df_baru_agg = df_baru_agg.rename(columns={'Harga': 'Harga_Baru'})

# Buat daftar unik nama mobil baru
list_mobil_baru = df_baru_agg['Mobil'].unique().tolist()

# Fungsi fuzzy matching nama mobil
def cocokkan_mobil(nama_mobil_bekas):
    hasil = process.extractOne(nama_mobil_bekas, list_mobil_baru, scorer=fuzz.token_sort_ratio)
    if hasil and hasil[1] >= 90:
        return hasil[0]
    else:
        return None

# Terapkan fuzzy matching
df_bekas['Mobil_Cocok'] = df_bekas['Mobil'].apply(cocokkan_mobil)

# Gabungkan berdasarkan nama mobil hasil pencocokan dan tahun
df_gabung = pd.merge(
    df_bekas,
    df_baru_agg,
    left_on=['Mobil_Cocok', 'Tahun'],
    right_on=['Mobil', 'Tahun'],
    how='left'
)

# Hitung depresiasi
df_gabung['Depresiasi_%'] = df_gabung.apply(
    lambda row: 100 * (row['Harga_Baru'] - row['Harga_Bekas']) / row['Harga_Baru']
    if pd.notnull(row['Harga_Baru']) and pd.notnull(row['Harga_Bekas']) else None,
    axis=1
)
df_gabung['Depresiasi_%'] = df_gabung['Depresiasi_%'].round(2)

# Rename kolom agar lebih informatif
df_gabung = df_gabung.rename(columns={
    'Mobil_x': 'Mobil_Bekas',
    'Mobil_y': 'Mobil_Baru'
})

# Tampilkan hasil
pd.options.display.float_format = '{:,.0f}'.format
display(df_gabung[['Mobil_Bekas', 'Mobil_Cocok', 'Tahun', 'Harga_Bekas', 'Harga_Baru', 'Depresiasi_%']].head(20))
```

```python
import numpy as np

# Buat dictionary depresiasi hanya dari tahun 20212025
df_valid = df_depresiasi_tahun[df_depresiasi_tahun['Tahun'] >= 2021]
depresiasi_dict = dict(zip(df_valid['Tahun'], df_valid['Depresiasi_%'] / 100))

# Default depresiasi hanya dihitung dari tahun 20212025
default_depresiasi = np.mean(df_valid['Depresiasi_%'] / 100)


# Tampilkan rata-rata depresiasi 20212025
print("Rata-rata depresiasi tahun 20212025 (sebagai default untuk tahun <2021):")
print(f"{default_depresiasi * 100:.2f}%")

# Salin dataframe
df = df_gabung.copy()
tahun_sekarang = 2025

# Fungsi estimasi harga baru berdasarkan depresiasi aktual
def estimasi_harga_baru(row):
    if pd.isna(row['Harga_Baru']):
        umur = tahun_sekarang - row['Tahun']
        harga_bekas = row['Harga_Bekas']

        depresiasi = depresiasi_dict.get(row['Tahun'], default_depresiasi)

        try:
            harga_baru = harga_bekas / ((1 - depresiasi) ** umur)
            return harga_baru
        except:
            return np.nan
    return row['Harga_Baru']

# Terapkan fungsi
df['Harga_Baru'] = df.apply(estimasi_harga_baru, axis=1)

# Hitung depresiasi % jika masih kosong
df['Depresiasi_%'] = df.apply(
    lambda row: ((row['Harga_Baru'] - row['Harga_Bekas']) / row['Harga_Baru']) * 100
    if pd.isna(row['Depresiasi_%']) else row['Depresiasi_%'],
    axis=1
)

# Lihat hasil
display(df[['Mobil_Bekas', 'Tahun', 'Harga_Bekas', 'Harga_Baru', 'Depresiasi_%']])
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

# Dataset dari df, pastikan kolom ini sudah tersedia
df_model = df[['Mobil_Bekas', 'Tahun', 'Harga_Baru', 'Harga_Bekas']].dropna().copy()

# Tambahkan kolom Umur Mobil
df_model['Umur_Mobil'] = 2025 - df_model['Tahun']

# Lakukan One-Hot Encoding untuk kolom nama mobil/merek mobil
df_model_encoded = pd.get_dummies(df_model, columns=['Mobil_Bekas'], drop_first=False)

# Fitur dan target
X = df_model_encoded[['Umur_Mobil'] + [col for col in df_model_encoded.columns if col != 'Harga_Bekas']]
y = df_model_encoded['Harga_Bekas']

# Normalisasi dengan MinMaxScaler untuk fitur dan target
scaler_X = MinMaxScaler()
X_scaled = scaler_X.fit_transform(X)

scaler_y = MinMaxScaler()
y_scaled = scaler_y.fit_transform(y.values.reshape(-1, 1))  # Reshape y untuk scaler

# Bagi data untuk training dan testing
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.2, random_state=42)
```

```python
print("X_train shape:", X_train.shape)
print("X_test shape :", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape :", y_test.shape)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam

# Membangun model Neural Network
model = Sequential()

# Input layer (jumlah fitur X_scaled)
model.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))

# Hidden layer 1
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.2))  # Regularisasi untuk mencegah overfitting

# Hidden layer 2
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.2))  # Regularisasi lagi

# Output layer (1 unit untuk harga mobil)
model.add(Dense(1))

# Kompilasi model
model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')

# Ringkasan model
model.summary()
```

```python
# Melatih model
history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_split=0.2, verbose=1)
```

```python
import numpy as np

# Prediksi harga mobil bekas pada data test
y_pred_scaled = model.predict(X_test)

# Kembalikan harga bekas ke skala semula menggunakan inverse_transform
y_pred = scaler_y.inverse_transform(y_pred_scaled)

# Kembalikan target y_test ke skala semula
y_test_rescaled = scaler_y.inverse_transform(y_test)

# Tampilkan hasil y_test (nilai asli) dan y_pred (nilai prediksi)
comparison_df = pd.DataFrame({
    'Y_Test': y_test_rescaled.flatten(),
    'Y_Pred': y_pred.flatten()
})

# Hitung MAPE
mape = np.mean(np.abs((y_test_rescaled - y_pred) / y_test_rescaled)) * 100

# Tampilkan perbandingan dan MAPE
display(comparison_df)
print(f"MAPE: {mape:.2f}%")
```
